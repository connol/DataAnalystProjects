### Question
Where should the company expand to maximize total possilbe profit?

```python
rg=rgdata.groupby('Region')
```

```python
rg['Total Profit'].aggregate(np.sum)
```

```python
prof=rgdata.groupby('Region')[['Total Profit', 'Region']].aggregate(np.sum).sort_values('Total Profit', ascending = False)
```

```python
plt.figure('Prof')
sns.barplot(y=prof.index, x='Total Profit', data = prof)
```

### Comment
As can be seen from the graph the Sub-Saharan Africa genereates the largest profit. A close second is the europe region. Based on an intial look the compnay should seek to expand in either Africa or Eurpoe. It may be useful to look at the costs to profit ratio to gain more insight


```python
pandc=rg[['Total Profit','Total Cost']].aggregate(np.sum)
```

```python
ratio= pandc['Total Profit']/pandc['Total Cost']
```

```python
pandc['Ratio']=ratio
pandc['Region']=pandc.index
```

```python
pandc.sort_values('Ratio',ascending = False ,inplace=True)
```

```python

plt.figure(figsize=(16,6))
plt.subplot(1,3,3)
sns.barplot(y='Region',x='Ratio',data=pandc)

plt.subplot(1,3,2)
sns.barplot(y='Region',x='Total Cost', data = pandc)

plt.subplot(1,3,1)
sns.barplot(y='Region', x='Total Profit', data = pandc)
plt.subplots_adjust(wspace=1)
```

### Conclusion
While profits are highest in Africa and Europe, the cost for both countries are also high. From out ratio chart we can see that each dollar spent in Asia generates more value then that in Europe and Africa by a small amount. We can also see that the cost in Asia is significantly lower as well. With a higher ratio and about half as much cost as Africa the company would find it beneficail to expand its Asia markets.

```python
Asd=rg.get_group('Asia').groupby('Country')[['Total Profit','Total Cost']]
```

```python
AA=Asd.aggregate(np.sum)
```

```python
AA["Ratio"]=Asd.aggregate(np.sum)['Total Profit']/Asd.aggregate(np.sum)['Total Cost']
AA['Country']=AA.index
```

```python
plt.figure(figsize=(20,10))
plt.subplot(1,3,1)
sns.barplot(y='Country',x = 'Total Profit', data = AA)

plt.subplot(1,3,2)
sns.barplot(y='Country', x = 'Total Cost', data = AA)

plt.subplot(1,3,3)
sns.barplot(y='Country', x = 'Ratio', data = AA)

plt.subplots_adjust(wspace=.5)
```

### Conclustion
Due to the high ratio of total profit to total cost along with an already low cost budget the company should seek to expand its markets in Kazakhstan to see an increase in profits.

### Question
What marketing strategy should the company use to increase profits?


```python
data.head()
```

```python
item_group=data.groupby('Item Type')
```

```python
item_group.count()
```

```python
item_intrests = item_group[['Total Profit','Total Cost','Units Sold','Unit Cost']].aggregate(np.sum)
```

```python
item_intrests
item_intrests['Item']= item_intrests.index
```

```python
lt = list()
for i in range(len(item_intrests)):
    lt.append(.1)
```

```python
plt.figure(figsize=(10,10))
colors = sns.color_palette("Paired")
plt.pie(item_intrests['Total Profit'],explode = lt ,autopct = '%1.1f%%',labels=item_intrests['Item'],colors=colors);
plt.title('Total Profit');
```

```python
plt.figure(figsize=(10,10))
colors = sns.color_palette("Paired")
plt.pie(item_intrests['Total Cost'],explode = lt ,autopct = '%1.1f%%',labels=item_intrests['Item'],colors=colors);
plt.title('Total Cost');
```

```python
plt.figure(figsize=(10,10))
colors = sns.color_palette("Paired")
plt.pie(item_intrests['Units Sold'],explode = lt ,autopct = '%1.1f%%',labels=item_intrests['Item'],colors=colors);
plt.title('Units Sold');
```

```python
rat = item_intrests['Total Profit']/item_intrests['Total Cost']
rat
item_intrests["Profit Ratio"]= rat
```

```python
fig, (a0,a1) = plt.subplots(1,2,figsize=(40,40))
plt.subplot(1,2,1)
colors = sns.color_palette("Paired")
plt.pie(item_intrests['Total Profit'],explode = lt ,autopct = '%1.1f%%',labels=item_intrests['Item'],colors=colors,textprops={'fontsize':20});
plt.title('Total Profit', fontsize= 20)

plt.subplot(1,2,2)
plt.pie(item_intrests['Total Cost'],explode = lt ,autopct = '%1.1f%%',labels=item_intrests['Item'],colors=colors, textprops={'fontsize':20});
plt.title('Total Cost', fontsize = 20);
```

```python
figsize=(10,10)
sns.barplot(x='Profit Ratio', y = 'Item', data = item_intrests);
plt.title('Profit Margin')
```

### Conclusion
While a large chunk of the company profits comes from cosmetics and household a large part of the company costs goes into making these products. Notice that clothes has a low cost percentage but a high profit which is backed up by the large profit margin as can be seen from the bar plot. Therefore, the company should look to increase there marketing and prodiction in clothes to increase there profits.
